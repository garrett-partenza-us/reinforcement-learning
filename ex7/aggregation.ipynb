```python
# Imports
import numpy as np
from IPython.display import clear_output
from tqdm import tqdm
import matplotlib.pyplot as plt
import random
```

```python
# Four rooms environment class copied and pasted from my ex0 submission
class Environment:
    
    def __init__(self, rows, cols, walls):
        
        self.rows = rows
        self.cols = cols
        self.walls = walls
        self.x = 0
        self.y = 0
        self.x_prev = 0
        self.y_prev = 0
        self.bank = 0
        
    def draw(self):
        for row in range(self.rows):
            for col in range(self.cols):
                if ((self.rows-row-1, col)) in self.walls:
                    print("{}   ".format("1"), end="")
                elif ((self.rows-row-1, col)) == (self.rows-1, self.cols-1):
                    print("{}   ".format("F"), end="")
                elif ((self.rows-row-1, col)) == (self.x, self.y):
                    print("{}   ".format("P"), end="")
                else:
                    print("{}   ".format("0"), end="")
            print("\n")
                        
    def step(self, action):
        slips = np.random.choice([True, False], p=[0.0, 1.0])
        self.x_prev=self.x
        self.y_prev=self.y
        if action == "LEFT":
            if slips:
                self.x=self.x+np.random.choice([1, -1], p=[0.5, 0.5])
            else:
                self.y=self.y-1
            self.correct()
            return self.reward()
        elif action == "RIGHT":
            if slips:
                self.x=self.x+np.random.choice([1, -1], p=[0.5, 0.5])
            else:
                self.y=self.y+1
            self.correct()
            return self.reward()
        elif action == "UP":
            if slips:
                self.y=self.y+np.random.choice([1, -1], p=[0.5, 0.5])
            else:
                self.x=self.x+1
            self.correct()
            return self.reward()
        elif action == "DOWN":
            if slips:
                self.y=self.y+np.random.choice([1, -1], p=[0.5, 0.5])
            else:
                self.x=self.x-1
            self.correct()
            return self.reward()
        
    def reward(self):
        if ((self.rows-self.x-1, self.y)) == (self.rows-1, self.cols-1):
            self.bank+=1
            self.restart()
            return (1, True)
        else:
            return (0, False)
        
    def correct(self):
        if self.x<0:
            self.x=self.x_prev
        elif self.x==self.rows:
            self.x=self.x_prev
        elif self.y<0:
            self.y=self.y_prev
        elif self.y==self.cols:
            self.y=self.y_prev
        elif ((self.x, self.y)) in self.walls:
            self.x=self.x_prev
            self.y=self.y_prev
            
    def restart(self):
        self.x = 0
        self.y = 0
        self.x_prev = 0
        self.y_prev = 0
        
    def reset(self):
        self.restart()
        self.bank = 0
        
    def loc(self):
        return (self.x, self.y)
```

```python
# Hyperparameters
action_space = ["UP", "DOWN", "LEFT", "RIGHT"]
ROWS = 11
COLS = 11
WALLS = [
#     (5, 0),
#     (5, 2),
#     (5, 3),
#     (5, 4),
#     (5, 5),
#     (4, 5),
#     (4, 6),
#     (4, 7),
#     (4, 9),
#     (4, 10),
#     (0, 5),
#     (2, 5),
#     (3, 5),
#     (6, 5),
#     (7, 5),
#     (9, 5),
#     (10, 5),
]
```

```python
def semi_grad_sarsa():
    
    #Input: a di↵erentiable action-value function parameterization
    
    q_features = {}
    for row in range(ROWS):
        for col in range(COLS):
            for a in action_space:
                features = np.array([
                    row,
                    col,
                    1 if ((row-1, col)) in WALLS else 0,
                    1 if ((row+1, col)) in WALLS else 0,
                    1 if ((row, col+1)) in WALLS else 0,
                    1 if ((row, col-1)) in WALLS else 0
                ])
                q_features[(row, col, a)] = features
                
    #initialize value-function weights
    w = {}
    for row in range(ROWS):
        for col in range(COLS):
            for a in action_space:
                w[(row, col, a)] = -np.random.rand(6)
                                
    #algorithm parameters
    alpha = 0.5
    epsilon = 0.2
    gamma = 1
    max_steps = 500
        
    #loop for each episode
    episodes = 500
    
    environment = Environment(rows=ROWS, cols=COLS, walls=WALLS)
    
    steps_to_finish = []
    
    for episode_num in tqdm(range(episodes)):
            
        term = False
                
        for step in range(max_steps):
        
            #initial state and action of episode
            x, y = environment.x, environment.y
            
            a = np.random.choice([max(action_space, key = lambda a: np.dot(w[row,col,a].T, q_features[x,y,a])), random.choice(action_space)], p=[1-epsilon, epsilon])
            r, term = environment.step(a)
            
            if term:
                target = np.dot(w[row,col,a].T, q_features[x,y,a]) + r
                value = np.dot(w[row,col,a].T, q_features[x,y,a])
                delta = alpha * (target - value)
                w[row,col,a] += float(delta)
                steps_to_finish.append(step)
                break
                
            else:
                
                x_new, y_new = environment.x, environment.y
                a = np.random.choice([max(action_space, key = lambda a: np.dot(w[row,col,a].T, q_features[x_new,y_new,a])), random.choice(action_space)], p=[1-epsilon, epsilon])
                target = gamma * (r + np.dot(w[row,col,a].T, q_features[x_new,y_new,a]) - np.dot(w[row,col,a].T, q_features[x,y,a]))
                value = np.dot(w[row,col,a].T, q_features[x,y,a])
                delta = alpha * (target - value)
                w[row,col,a] += float(delta)
                
        environment.reset()
        
    return steps_to_finish
                                
stf = semi_grad_sarsa()
```
